Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Anomaly Detection

Anomaly Detection is the process of identifying data points, events, or patterns that do not follow the normal behavior of a dataset. These unusual observations are called anomalies or outliers and often indicate important events such as fraud, system failures, or medical issues.

Types of Anomalies
1. Point Anomaly

A point anomaly is a single data point that is significantly different from the rest of the data.

Example:

In a bank transaction dataset, most withdrawals are between ‚Çπ1,000 and ‚Çπ10,000.

A single transaction of ‚Çπ5,00,000 appears.

This single value is a point anomaly and may indicate fraud.

üìå Simple example:
A student usually scores between 60‚Äì80 marks, but suddenly scores 5 marks in one exam.

2. Contextual Anomaly

A contextual anomaly is a data point that is abnormal only in a specific context (such as time, place, or condition).

Example:

A temperature of 30¬∞C is normal during summer.

The same 30¬∞C in winter is unusual.

Hence, it becomes a contextual anomaly.

üìå Simple example:
High electricity usage at night in a closed office building.

3. Collective Anomaly

A collective anomaly occurs when a group of related data points is anomalous, even if individual points seem normal.

Example:

A single network request is normal.

But hundreds of requests in a short time from the same IP address indicate a DDoS attack.

The group together forms a collective anomaly.

üìå Simple example:
A heart rate suddenly shows rapid irregular beats continuously‚Äîthe pattern is abnormal, not each beat alone.

Summary Table
Type of Anomaly	Description	Example
Point	Single unusual data point	One huge bank withdrawal
Contextual	Unusual in a specific context	High temperature in winter
Collective	Group of unusual data points	Multiple failed login attempts

‚úÖ In short:

Point anomaly ‚Üí One strange value

Contextual anomaly ‚Üí Strange due to situation

Collective anomaly ‚Üí Strange as a group

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) üöÄ

These are popular unsupervised anomaly detection techniques, but they work in very different ways and are useful in different situations.

1. Isolation Forest (iForest) üå≥
Approach

Isolates anomalies instead of profiling normal data.

Uses random splits in decision trees.

Anomalies get isolated faster because they are rare and different.

Key Idea

üëâ ‚ÄúAnomalies are few and different, so they are easier to separate.‚Äù

Suitable Use Cases

Large datasets üìä

High-dimensional data

Fraud detection üí≥

Network intrusion detection üåê

Pros & Cons

‚úÖ Fast and scalable
‚úÖ Works well with many features
‚ùå Less effective for local anomalies

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) üîµ
Approach

Groups data points based on density.

Points in low-density regions are labeled as outliers.

Does not require the number of clusters in advance.

Key Idea

üëâ ‚ÄúNormal points form dense clusters; anomalies lie in sparse areas.‚Äù

Suitable Use Cases

Spatial data üìç

Data with clusters of varying shapes

Noise detection

GPS and geolocation data üó∫Ô∏è

Pros & Cons

‚úÖ Identifies noise naturally
‚úÖ Finds arbitrarily shaped clusters
‚ùå Sensitive to parameters (eps, min_samples)
‚ùå Struggles in high dimensions

3. Local Outlier Factor (LOF) üìâ
Approach

Compares local density of a point with its neighbors.

A point is an outlier if it has much lower density than nearby points.

Key Idea

üëâ ‚ÄúA point is anomalous if it is isolated compared to its neighbors.‚Äù

Suitable Use Cases

Detecting local anomalies

Small to medium datasets

Data with varying densities

Sensor data ‚öôÔ∏è

Pros & Cons

‚úÖ Excellent for local outliers
‚úÖ Adapts to varying densities
‚ùå Computationally expensive
‚ùå Not ideal for very large datasets

Question 3: What are the key components of a Time Series? Explain each with one example.


Question 3: Key Components of a Time Series

A Time Series is data collected over time at regular intervals (daily, monthly, yearly).

Key Components
1. Trend

The long-term upward or downward movement in data.

Example:
Increasing number of mobile users over years üìà

2. Seasonality

Patterns that repeat at fixed intervals (daily, monthly, yearly).

Example:
Higher ice-cream sales every summer üç¶

3. Cyclic Component

Long-term fluctuations without fixed periods.

Example:
Economic growth and recession cycles.

4. Irregular (Random) Component

Unpredictable, sudden changes.

Example:
Drop in stock prices due to natural disasters or news events.

Summary Table
Component	Meaning	Example

Seasonality	Repeating pattern	Festival sales

‚úÖ In short:

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Question 4: Stationarity in Time Series
What is a Stationary Time Series?

A stationary time series is one whose statistical properties remain constant over time, such as:

Mean

Variance

Covariance

üìå Why important?
Most forecasting models (like ARIMA) work only on stationary data.

Example

Stationary: Daily temperature fluctuates around the same average.

Non-stationary: Population growth increasing every year.

How to Test Stationarity
1. Visual Inspection

Plot the data

If mean or variance changes ‚Üí non-stationary

2. Augmented Dickey-Fuller (ADF) Test

Null hypothesis: Data is non-stationary

p-value < 0.05 ‚Üí Stationary

How to Convert Non-Stationary to Stationary
üîπ 1. Differencing

Subtract current value from previous value.

Example:
Y(t) ‚Äì Y(t-1)

üîπ 2. Log Transformation

Reduces large variations.

Example:
Useful for exponential growth data.

üîπ 3. Remove Trend & Seasonality

Using decomposition methods.

‚úÖ Summary

Stationary ‚Üí constant mean & variance

Test ‚Üí ADF test

Transform ‚Üí Differencing, log, detrending

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
)

# Convert Month to datetime
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Plot original time series
plt.figure()
plt.plot(data['Passengers'])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Decompose the series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

Interpretation

Trend: Overall growth in air passengers ‚úàÔ∏è

Seasonality: Repeating yearly pattern (summer peaks)

Residual: Random fluctuations

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Synthetic taxi fare dataset
np.random.seed(42)
fare_amount = np.concatenate([np.random.normal(15, 3, 300), np.random.normal(60, 5, 20)])
trip_distance = np.concatenate([np.random.normal(3, 1, 300), np.random.normal(20, 3, 20)])

df = pd.DataFrame({
    'fare_amount': fare_amount,
    'trip_distance': trip_distance
})

# Isolation Forest
iso = IsolationForest(contamination=0.06, random_state=42)
df['anomaly'] = iso.fit_predict(df)

# Visualization
plt.figure()
plt.scatter(df['fare_amount'], df['trip_distance'], c=df['anomaly'])
plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("Isolation Forest Anomaly Detection")
plt.show()
```

Interpretation

Normal points ‚Üí Regular taxi trips

Anomalies ‚Üí Very high fare + long distance trips (possible fraud or errors)

‚úÖ Summary

Isolation Forest ‚Üí Fast, scalable anomaly detection

Time Series Components ‚Üí Trend, Seasonality, Residual

Stationarity ‚Üí Required for ARIMA models

SARIMA/SARIMAX ‚Üí Seasonal forecasting

Decomposition + Isolation Forest ‚Üí Practical real-world tools

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(data,
                order=(1,1,1),
                seasonal_order=(1,1,1,12))
results = model.fit()

forecast = results.forecast(steps=12)

plt.plot(data, label='Original')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.title('SARIMA Forecast - Next 12 Months')
plt.show()
```

üìå Used for seasonal forecasting (monthly airline data)

‚úÖ Summary

Anomaly Detection ‚Üí Isolation Forest

Seasonal Time Series ‚Üí SARIMA

Non-stationary data ‚Üí Differencing

Real-world forecasting ‚Üí SARIMAX

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python
from sklearn.neighbors import LocalOutlierFactor

X = np.random.normal(0, 1, (300, 2))
lof = LocalOutlierFactor(n_neighbors=20)
labels = lof.fit_predict(X)

plt.figure()
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Local Outlier Factor Anomalies")
plt.show()
```
